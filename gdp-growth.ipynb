```python
import matplotlib.pyplot as plt

# Data for GDP per capita
years = [1950, 1955, 1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020]
japan_gdp = [3062, 4417, 6354, 9459, 15484, 18082, 21404, 24437, 29949, 31887, 33211, 34858, 35011, 37053, 36827]
china_gdp = [799, 1119, 1057, 1253, 1398, 1594, 1930, 2619, 2982, 4000, 4730, 6627, 9658, 13325, 17226]

# Calculate ratio to 1950 GDP
china_base = china_gdp[0]
japan_base = japan_gdp[0]
china_ratio = [gdp / china_base for gdp in china_gdp]
japan_ratio = [gdp / japan_base for gdp in japan_gdp]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(years, china_gdp, label="China", marker="o")
plt.plot(years, japan_gdp, label="Japan", marker="o")
plt.title("GDP per Capita for China and Japan (Levels)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("GDP per Capita (Levels)", fontsize=12)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(years, china_ratio, label="China (Ratio to 1950)", marker="o")
plt.plot(years, japan_ratio, label="Japan (Ratio to 1950)", marker="o")
plt.title("GDP per Capita as Ratio to 1950 Levels", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Ratio to 1950 Level", fontsize=12)
plt.legend()
plt.grid()
plt.show()
```

```python

data = {
    "Country": ["United States", "Canada", "France", "Hong Kong", "South Korea", "Indonesia", "Argentina", "Mexico", "Kenya", "Ethiopia"],
    "Capital per Person": [200541, 173316, 159651, 205239, 176910, 31436, 39824, 49357, 5288, 3457],
    "Per Capita GDP": [62589, 49884, 43755, 54810, 42219, 11595, 21827, 18737, 4237, 2721],
}

import pandas as pd

df = pd.DataFrame(data)


us_capital = df.loc[0, "Capital per Person"]
us_gdp = df.loc[0, "Per Capita GDP"]
df["Capital per Person (relative)"] = df["Capital per Person"] / us_capital
df["Per Capita GDP (relative)"] = df["Per Capita GDP"] / us_gdp


df["Predicted y*"] = df["Capital per Person (relative)"] ** (1 / 3)


df["Implied TFP"] = df["Per Capita GDP (relative)"] / df["Predicted y*"]


df
```
